# Matplotlib

#### Load necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic graph

```python
x = [0,1,2,3,4]
y = [0,1.5,2,3.,4.5]
plt.plot(x,y)
plt.title('Siple explo of Graph', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 10})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,1.5,2,3.,4.5])
plt.show()
```
