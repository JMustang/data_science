# Loading data fom the fifa`s game

```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading data
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

#### Histograms

```python
plt.figure(figsize=(13,7))

bins = [40,50,60,70,80,90,100]

plt .hist(fifa.Overall, bins=bins, color='green')
plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Dristribution of player skills in FIFA 2010',fontdict={'fontweight':'bold', 'fontsize':20})

plt.show()
```
