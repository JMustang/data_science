# Loading data from the fifa`s game

```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Loading data
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

#### Histograms

```python
plt.figure(figsize=(13,7))

bins = [40,50,60,70,80,90,100]

plt .hist(fifa.Overall, bins=bins, color='green')
plt.xticks(bins)
plt.yticks([0, 2000, 4000, 6000, 8000, 10000], ["0",'2K','4K','6K','8K','10K',])

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Dristribution of player skills in FIFA 2010',fontdict={'fontweight':'bold', 'fontsize':20})

plt.show()
```

#### Pie Charts

```python
# Preferred kick with left foot
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
left
```

```python
# Preferred kick with right foot
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
right
```

```python
labels = ['Left Foot', 'Right Foot']
colors = ['red', 'green']

plt.style.use('ggplot')

plt.pie([left, right], labels=labels, colors=colors, autopct='%.2f %%')
plt.title('Foot Preference of FIFA Players', fontdict={'fontweight':'bold', 'fontsize':20, 'color':'gray'})
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]


weights = [light, light_medium, medium, medium_heavy, heavy]
lbls = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.1,0,0,.4)

plt.pie(weights, labels=lbls, autopct='%.2f %%', pctdistance=0.8, explode=explode)
plt.title('Weights Distribution of FIFA Players (in lbs)', fontdict={'fontweight':'bold', 'fontsize':20, 'color':'gray'})
plt.show()
```

#### Box and whisker plot

```python
plt.style.use('default')

plt.figure(figsize=(6, 8))

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
liverpool = fifa.loc[fifa.Club == 'Liverpool']['Overall']
manchester = fifa.loc[fifa.Club == 'Manchester United']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'Liverpool', 'Manchester United']

boxes = plt.boxplot([barcelona, madrid, liverpool, manchester], labels=labels, patch_artist=True, medianprops={'linewidth':2})

for box in boxes['boxes']:
  # Setting the edge color
  box.set(color='#4286f4', linewidth=2)

  # change fill color
  box.set(facecolor='#e0e0e0')

plt.title('Professional soccer team comparison')
plt.ylabel('FIFA Overall Rating')

plt.show()
```
