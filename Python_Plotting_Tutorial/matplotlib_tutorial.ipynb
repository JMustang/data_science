# Real World Examples

```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Line Graph

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas prices over time (in USD)')

plt.plot(gas.Year, gas.USA,'.-', color="red", label='USA')
plt.plot(gas.Year, gas.Canada,'^-', color='green', label='Canada')
plt.plot(gas['Year'], gas['South Korea'],'.-', color='blue', label='South Korea')

plt.xticks(gas.Year, rotation='vertical', size=10)

plt.xlabel('YEAR')
plt.ylabel('US DOLAR')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(13,7))

plt.title('Gas prices over time (in USD)', fontdict={'fontweight':'bold', 'fontsize':20})

for country in gas:
  if country != 'Year':
    plt.plot(gas.Year, gas[country], marker='.', label=country)

plt.xticks(gas.Year[::3])
plt.yticks([1,2,3,4,5,6,7,8], ['1.B','2.B','3.B','4.B','5.B','6.B','7.B','8.B'])

plt.xticks(gas.Year, rotation='vertical', size=10)

plt.xlabel('YEAR')
plt.ylabel('US DOLAR')

plt.legend(fontsize=9)
plt.savefig('Gas_price_figure.png', dpi=300)
plt.show()
```
